```python
import pandas as pd

# Nama file CSV yang muncul saat upload
df = pd.read_csv('Data/Data Jurusan Peminat Saintex.csv')

# Lihat 5 data pertama
df.head()
```

### 2.Eksplorasi data dan Bersihkan Data

```python
# Cek kolom yang tersedia
df.columns

# Drop kolom kosong (Unamed)
df = df.loc[:, ~df.columns.str.contains('^Unnamed')]

# Rename kolom agar lebih mudah diakses
df.columns = df.columns.str.strip().str.lower().str.replace(" ", "_")
df.rename(columns={
    "nama_ptn": "asal_univ",
    "jurusan": "nama",
    "daya_tampung_(tahun_2025)": "daya_tampung_2025",
    "jumlah_peminat": "peminat"
}, inplace=True)

df.head()
```

### 3. Analisis Umum: Jumlah Program Studi, PTN, dan Total Daya Tampung

```python
jumlah_jurusan_unik = df['nama'].nunique()
jumlah_univ_unik = df['asal_univ'].nunique()
total_daya_tampung = df['daya_tampung_2025'].sum()

print(f"Jumlah Jurusan Unik: {jumlah_jurusan_unik}")
print(f"Jumlah PTN: {jumlah_univ_unik}")
print(f"Total Daya Tampung 2025: {total_daya_tampung}")
```

### 4. Visualisasi Jumlah Peminat per Jurusan

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_jurusan = df.groupby('nama')['peminat'].sum().sort_values(ascending=True).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_jurusan.values, y=top_jurusan.index, palette='viridis')
plt.title("Jurusan dengan Peminat Terendah")
plt.xlabel("Jumlah Peminat")
plt.ylabel("Jurusan")
plt.show()
```

### 5. Visualisasi Daya Tampung per Jurusan

```python
top_daya = df.groupby('nama')['daya_tampung_2025'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_daya.values, y=top_daya.index, palette='magma')
plt.title("Top 10 Jurusan dengan Daya Tampung Tertinggi (2025)")
plt.xlabel("Daya Tampung")
plt.ylabel("Jurusan")
plt.show()
```

### 6. Visualisasi Pie Chart Jenjang Pendidikan

### 7. Visualisasi Perbandingan Peminat dengan Daya Tampung

```python
top_8 = df.groupby('nama')[['peminat', 'daya_tampung_2025']].sum().sort_values(by='peminat', ascending=False).head(8)
bottom_8 = df.groupby('nama')[['peminat', 'daya_tampung_2025']].sum().sort_values(by='peminat', ascending=True).head(8)

def plot_dual_bar(data, title):
    data.plot(kind='bar', figsize=(12,6))
    plt.title(title)
    plt.xlabel("Jurusan")
    plt.ylabel("Jumlah")
    plt.xticks(rotation=45)
    plt.legend(["Peminat", "Daya Tampung"])
    plt.grid(True)
    plt.tight_layout()
    plt.show()

plot_dual_bar(top_8, "Top 8 Jurusan: Peminat dengan Daya Tampung")
plot_dual_bar(bottom_8, "8 Jurusan: Peminat dengan Daya Tampung")
```

### 8 Visualisasi Gaji tertinggi menurut jurusan

```python
# Nama kolom sesuai dataset asli
if 'rata-rata_gaji_lulusan' in df.columns:
    # Membersihkan nilai agar menjadi numerik (jika perlu), hapus karakter non-numerik
    df['gaji_bersih'] = df['rata-rata_gaji_lulusan'].replace('[^0-9]', '', regex=True).astype(float)

    # Hapus duplikat berdasarkan kolom 'nama' (jurusan), ambil hanya gaji tertinggi per jurusan
    data_filtered = df.loc[df.groupby('nama')['gaji_bersih'].idxmax()]

    # Tampilkan 10 jurusan dengan nilai gaji tertinggi
    top_gaji_langsung = data_filtered[['nama', 'gaji_bersih']].sort_values(by='gaji_bersih', ascending=False).head(10)
    print("Top 10 Gaji Tertinggi (Tanpa Rata-Rata):")
    print(top_gaji_langsung)

    # Visualisasi
    plt.figure(figsize=(12, 6))
    sns.barplot(x='gaji_bersih', y='nama', data=top_gaji_langsung, palette='viridis')
    plt.title("Top 10 Gaji Tertinggi Tanpa Rata-Rata")
    plt.xlabel("Gaji (Rp)")
    plt.ylabel("Nama Jurusan")
    plt.tight_layout()
    plt.show()

else:
    print("Kolom 'rata-rata_gaji_lulusan' tidak ditemukan. Mohon pastikan nama kolom yang benar.")
```

### 9. Visualisasi Heatmap Peminat per Provinsi#




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# kolom yang digunakan ada dalam data
if 'provinsi' in df.columns and 'peminat' in df.columns:

    # Menghitung total peminat per provinsi
    provinsi_data = df.groupby('provinsi')['peminat'].sum()

    # Mengubah data menjadi DataFrame untuk keperluan visualisasi
    provinsi_data = provinsi_data.reset_index()

    # Menampilkan data peminat per provinsi untuk memastikan tidak ada kesalahan
    print(provinsi_data.head())

    # Membuat heatmap untuk visualisasi
    # Mengubah DataFrame ke bentuk pivot agar bisa di-heatmap
    heatmap_data = provinsi_data.set_index('provinsi').T  # Transformasi data menjadi bentuk pivot

    plt.figure(figsize=(12, 8))
    sns.heatmap(heatmap_data, annot=True, cmap='YlGnBu', fmt=",.0f", cbar_kws={'label': 'Jumlah Peminat'})
    plt.title('Heatmap Peminat per Provinsi')
    plt.xlabel('Provinsi')
    plt.ylabel('Jumlah Peminat')
    plt.tight_layout()
    plt.show()

else:
    print("Kolom 'provinsi' atau 'peminat' tidak ditemukan.")
```

### 10. Visualisasi Scatter Plot: Rasio Keketatan dengan Rata-rata Gaji Lulusan#

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Memastikan dataframe bernama df dan kolomnya sudah bersih
df_scatter = df[['rasio_keketatan', 'rata-rata_gaji_lulusan', 'nama']].dropna()

# Konversi jika ada nilai string atau koma
df_scatter['rasio_keketatan'] = pd.to_numeric(df_scatter['rasio_keketatan'], errors='coerce')
df_scatter['rata-rata_gaji_lulusan'] = pd.to_numeric(df_scatter['rata-rata_gaji_lulusan'], errors='coerce')
df_scatter = df_scatter.dropna()

# Membuat scatter plot
plt.figure(figsize=(12, 7))
sns.scatterplot(
    data=df_scatter,
    x='rasio_keketatan',
    y='rata-rata_gaji_lulusan',
    hue='nama',  # Opsional: warnai berdasarkan nama jurusan
    palette='tab10',
    legend=False,
    s=100,
    alpha=0.7
)

plt.title('Scatter Plot: Rasio Keketatan vs Rata-rata Gaji Lulusan', fontsize=14)
plt.xlabel('Rasio Keketatan')
plt.ylabel('Rata-rata Gaji Lulusan (Rp)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
